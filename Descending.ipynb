```python
def find_max(lst):
    max_val = lst[0]
    for i in lst:
        if i > max_val:
            max_val = i
    return max_val
def sort_list_desc(lst):
    sorted_list = []
    while lst:
        max_val = find_max(lst)
        lst.remove(max_val)
        sorted_list.append(max_val)
    return sorted_list
L1 = [80, 1, 6, 101, 8, 3]
L2 = sort_list_desc(L1)
print("Sorted list in descending order:", L2)
```
